## 1. Loading packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 2. Introduction to Numpy

```python
# np.genfromtxt() reads text data
data = np.genfromtxt('Atlanta_Monthly_Mean_Temperature_NWS.csv',delimiter=',',skip_header=2)
```

```python
# .dtype tells you what data type it is in
data.dtype
```

```python
# .shape tells you the dimension of numpy array
data.shape
```

```python
# get year from the first column
year = data[:,0]
# get July data from the 8th column (index = 7)
Jul = data[:,7]
```

```python
# simply display the data
print('year is ',year)
print('July temperature is ',Jul)
```

## 3. Introduction to matplotlib

```python
# generate a time series plot
plt.plot(year,Jul)
# add x label
plt.xlabel('time')
# add y label
plt.ylabel('monthly mean July temperature, deg F')
# add title
plt.title('monthly mean temperature of Atlanta GA')
# if preferred, you can generate a graphic file
plt.savefig('Figure1-1.pdf',bbox_inches='tight')
plt.show()
```

```python
# Generate a histogram of July temperature
plt.hist(Jul)
# add x label
plt.xlabel('monthly mean July temperature, deg F')
# add y label
plt.ylabel('frequency')
# add title
plt.title('monthly mean temperature of Atlanta GA')
# if preferred, you can generate a graphic file
plt.savefig('Figure1-2.pdf',bbox_inches='tight')
plt.show()
```
